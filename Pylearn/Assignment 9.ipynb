```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
data = pd.read_csv("data.csv", dtype=str, sep=',')
t=data['日期']
x=data.index
y1=list(map(int,data['全国累计确诊人数']))
y2=list(map(int,data['全国每日新增确诊']))
plt.figure(figsize=(15,6))
plt.plot(x, y1,label='全国累计确诊人数')
plt.legend(loc=1,ncol=2)
plt.xticks(x, t)
plt.xticks(rotation=90) 
plt.twinx()
plt.plot(x, y2,'r',label='全国每日新增确诊')
plt.legend(loc=4,ncol=2)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
data = pd.read_csv("data.csv", dtype=str, sep=',')
t=data['日期']
plt.figure(figsize=(15,10))
x=data.index
y=list(map(int,data['全国每日新增确诊']))
bubble =plt.scatter(x,y,c=y,cmap='Reds',s=[i/5 for i in y])
plt.colorbar(bubble)
plt.xticks(x, t)
plt.xticks(rotation=90) 
plt.title('全国每日新增确诊')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
data = pd.read_csv("data.csv", dtype=str, sep=',')
y=list(map(int,data['全国每日新增确诊']))
w=np.zeros(4)
for i in range(4):
    s=0
    for j in range(7*i,7*(i+1)):
        s+=y[j+3]
    w[i]=s
plt.figure(figsize=(5,5))
plt.pie(w,labels=['week 1','week 2','week 3','week 4'],pctdistance=0.7,autopct='%.1f%%',wedgeprops=dict(edgecolor='w', width=0.7, linewidth=10))
plt.show()
```
