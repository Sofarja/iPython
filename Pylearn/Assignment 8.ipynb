```python
import turtle as t
from turtle import *
from IPython.display import Image
from PIL import Image as IMG
path=r'C:\\Users\87947\Desktop\iPython\Pylearn\A8P\circle'
t.setup(500, 500)
colorList = ["red","pink","black","yellow","black","yellow","blue","yellow","yellow"]
for i in range(1,10):
    t.penup()
    t.sety(-15*(10-i)) 
    t.pendown()
    t.begin_fill()
    t.color(colorList[i-1])
    t.pencolor("black")
    t.circle(15*(10-i))
    t.end_fill()
p = t.getscreen()
p.getcanvas().postscript(file=path+".eps")
t.done()

im = IMG.open(path+".eps")
im.save(path+".jpg", "JPEG")
Image(path+".jpg", width=500, height=500)
```

```python
import turtle as t
from turtle import *
from IPython.display import Image
from PIL import Image as IMG


def koch(size,n):
    if n==0:
        t.forward(size)
    else:
        for angle in [0,60,-120,60]:
            t.left(angle)
            koch(size/3,n-1)

path=r'C:\\Users\87947\Desktop\iPython\Pylearn\A8P\koch'
t.setup(500, 500)
t.speed(10)
t.penup()
t.goto(-200, 100)
t.pendown()
t.pensize()
level=4
for i in range(3):
    koch(400, level)
    t.right(120)
    
p = t.getscreen()
p.getcanvas().postscript(file=path+".eps")
t.done()

im = IMG.open(path+".eps")
im.save(path+".jpg", "JPEG")
Image(path+".jpg", width=500, height=500)
```
