```python
import numpy as np

def pivot():
    l = list(d[0][:-1])
    print(l)
    print(d[0])
    jnum = l.index(max(l)) #转入下标
    print("转入下标",jnum)
    m = []
    for i in range(bn):
        if d[i][jnum] == 0:
            m.append(float('inf'))
        else:
            m.append(d[i][-1]/d[i][jnum])
    print("m",m)
    θ=[x for x in m[1:]]
    for i in range(bn-1):
        if θ[i]<=0:
            θ[i]=float('inf')
    print("theta",θ)
    inum = m.index(min(θ))  #转出下标
    print("转出下标",inum)
    s[inum-1] = jnum
    print("基变量列表",s)
    r = d[inum][jnum]
    d[inum] /= r
    for i in [x for x in range(bn) if x !=inum]:
        r = d[i][jnum]
        d[i] -= r * d[inum]
        
def solve():
    flag = True
    while flag:
        if max(list(d[0][:-1])) <= 0: #直至所有系数小于等于0
            flag = False
        else:
            pivot()
            
def printSol():
    for i in range(cn - 1):
        if i in s:
            print("x"+str(i+1)+"=%.2f" % d[s.index(i)+1][-1])
        else:
            print("x"+str(i+1)+"=0.00")
    print("objective is %.2f"%(-d[0][-1]))
```

```python
d = np.loadtxt("data.txt", dtype=np.float)
print(d)
(bn,cn) = d.shape
print(d.shape)
s = list(range(cn-bn,cn-1)) #基变量列表
print(s)

for j in range(cn-1,cn):#大M法目标值预处理
    S=0
    for i in range(1,bn):
        S=S+d[i][j]*d[0][cn-bn+i-1]
    print(d[0][j],S)
    d[0][j]=d[0][j]-S

for j in range(cn-1):#大M法目标函数预处理
    S=0
    for i in range(1,bn):
        S=S+d[i][j]*d[0][cn-bn+i-1]
    print(d[0][j],S)
    d[0][j]=d[0][j]-S
    
print(d)
solve()
printSol()
```
