```python
#！/usr/bin/env python
# -*- coding:utf-8 -*-
import pandas as pd

import time

def read_csv():
    date = '20170102'
    csv_path = open('MET'+ date + '.csv',encoding = 'utf-8')
    filepath_1 = '统计' + date + '进站客流.txt'
    text_df = pd.read_csv(csv_path,dtype = str,sep = ' ')
    text_df.notnull()
    text_df.dropna(how="any")
    print("正在转换格式，请稍等...")
    start_1 = time.time()
    text_df['入站时间'] = pd.to_datetime(text_df['入站时间'])
    #print(text_df['入站时间'])
    end_1 = time.time()
    print("转换完毕，用时%.2f秒" % (end_1 - start_1))

    print("正在统计，请稍等...")
    start_2= time.time()
    station_ID = 9

    every_number = 0
    station_total = 0
    hour_1 = 3
    min_1 = 0
    hour_2 = 3
    min_2 = 9

    hour_1 = str(hour_1)
    min_1 = str(min_1)
    hour_2 = str(hour_2)
    min_2 = str(min_2)

    f = open(filepath_1, 'w')
    while int(hour_2) < 24:
        station_volume = 0
        data = text_df[(text_df['入站时间'] >= pd.to_datetime(date + ' ' + hour_1 + ':' + min_1 + ':00', format='%Y-%m-%d %H:%M:%S')) & (
                            text_df['入站时间'] < pd.to_datetime(date + ' ' + hour_2 + ':' + min_2 + ':59', format='%Y-%m-%d %H:%M:%S'))]
        every_number += len(data)
        for num in data['入站站点号']:
            if int(num) == int(station_ID):
                station_volume += 1
        station_total += station_volume
        pre_write = date + ' ' + hour_1 + ':' + min_1 + ':00' + '  ' + str(station_volume) + ' ' + str(station_total) + '\n'
        f.write(pre_write)

        hour_1 = int(hour_1)
        hour_2 = int(hour_2)
        min_1 = int(min_1)
        min_2 = int(min_2)
        min_1 += 10
        min_2 += 10
        if min_1 > 50:
            min_1 = 0
            hour_1 += 1
        if min_2 > 60:
            hour_2 += 1
            min_2 = 9

        hour_1 = str(hour_1)
        min_1 = str(min_1)
        hour_2 = str(hour_2)
        min_2 = str(min_2)

    end_2 = time.time()
    print("统计完毕，用时%.2f秒" % (end_2 - start_2))

if __name__ == '__main__':
    read_csv()
```
