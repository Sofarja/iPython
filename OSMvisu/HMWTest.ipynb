```python
import folium
import folium.plugins as plugins
import numpy as np

np.random.seed(3141592)
initial_data = (
    np.random.normal(size=(100, 2)) * np.array([[1, 1]]) +
    np.array([[48, 5]])
)

move_data = np.random.normal(size=(100, 2)) * 0.01

data = [(initial_data + move_data * i).tolist() for i in range(100)]

weight = 1  # default value
for time_entry in data:
    for row in time_entry:
        row.append(weight)
```

```python
m = folium.Map([48., 5.], tiles='stamentoner', zoom_start=6)

hm = plugins.HeatMapWithTime(data)

hm.add_to(m)

m
```

```python
from datetime import datetime, timedelta


time_index = [
    (datetime.now() + k * timedelta(1)).strftime('%Y-%m-%d') for
    k in range(len(data))
]
```

```python
import branca
import os

m = folium.Map([48., 5.], tiles='stamentoner', zoom_start=6)

hm = plugins.HeatMapWithTime(
    data,
    index=time_index,
    auto_play=True,
    max_opacity=0.3
)

# create a Figure, with a map inside.
f = branca.element.Figure()
h=folium.Map([48., 5.], tiles='stamentoner', zoom_start=6)
hm.add_to(h)
h.add_to(f)

# put the figure into an IFrame.
iframe = branca.element.IFrame(width=500, height=300)
f.add_to(iframe)

# put the IFrame in a Popup
popup = folium.Popup(iframe, max_width=2650)

# create another map.
m = folium.Map([48., 5.], zoom_start=4)


# put the Popup on a marker, in the second map.
folium.Marker([48., 5.], popup=popup).add_to(m)
hm2 = plugins.HeatMapWithTime(
    data,
    index=time_index,
    auto_play=True,
    max_opacity=0.3
)
hm2.add_to(m)
# get a map in a Popup.
m.save(os.path.join('popups.html'))

m
```
