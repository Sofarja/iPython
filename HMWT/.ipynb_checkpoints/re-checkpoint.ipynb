```python
import pandas as pd
import numpy as np
data=pd.read_excel("数据汇总++.xlsx")
print(data)
print(data['车辆牌照'][0])
```

```python
from collections import Counter
num = dict(Counter(data['车辆牌照']))
print(num)
```

```python
import numpy as np
import pandas as pd
import xlwd
writer = pd.ExcelWriter('A.xlsx')

IDlist=[]
Trailist=[]
for ID in data['车辆牌照']:
    if num[ID]>1 and ID not in IDlist:
        flag=0
        C=[]
        IDlist.append(ID)
        check=np.array(data[:][data['车辆牌照']==ID])
        
        for c in check:
            C.extend(np.split(c,1))
        time=[]
        for i in range(4,74):
            t=[]
            for c in C:
                t.append(c[i])
            time.append(t)
        for t in time:
            if Counter(t)[0]>1:
                flag=1
        if flag==1:
            Trailist.append(C)
```

```python
import xlwt
# 创建一个workbook 设置编码
workbook = xlwt.Workbook(encoding = 'utf-8')
# 创建一个worksheet
worksheet = workbook.add_sheet('My Worksheet')

# 写入excel
# 参数对应 行, 列, 值
worksheet.write(1,0, label = 'this is test')

# 保存
workbook.save('Excel_test.xls')
```
